In this report, we delve into a dataset containing Twitter data to uncover insights about retweet counts, popular and unpopular dog breeds, and high-rated dog breeds, according to WeRateDogs. The dataset provides a fascinating glimpse into the world of social media and dog lovers. Let's dive into the findings!

## Average Retweet Count for Different Source Platforms:

 

We began by examining the tweet count for different source platforms. The majority of tweets(98%) were published from iPhone

	

Our analysis revealed exciting trends. Among the popular platforms, Twitter for iPhone emerged as the clear leader with an average retweet count of 2931. TweetDeck followed closely behind with an average of 2599 retweets. Surprisingly, the Twitter Web Client had a comparatively lower average retweet count of 2441. These findings highlight the varying engagement levels across different source platforms.
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 

                                                         
## Popular and Unpopular Dog Breeds:

Next, we turned our attention to dog breeds. 


Based on the dataset, we aimed to identify the most popular and unpopular breeds. Our analysis uncovered some intriguing results. Among the unpopular breeds, we found breeds such as Whippet(avg 1280), Redbone(avg 1175), Miniature Schnauzer(avg 763), English Springer(avg 763), and Giant Schnauzer(avg 739). These breeds had relatively fewer mentions or interactions in the dataset. On the other hand, the popular breeds included Black-and-Tan Coonhound(avg  33911), Leonberg(avg  29684), French Bulldog(avg  23692), Cardigan(avg 21346), and Great Pyrenees(avg 21109). These breeds garnered significant attention and engagement from the Twitter community.


## High-Rated Dog Breeds according to WeRateDogs:

Lastly, we sought to determine the most high-rated dog breeds according to the famous WeRateDogs Twitter account. 


Our analysis focused on the rating numerator assigned to each breed. The breeds with the highest ratings were Gordon Setter and Black-and-Tan Coonhound, both receiving a rating of 14/10. Afghan Hound, Malinois, and Leonberg also earned high ratings of 13/10. These breeds are undoubtedly cherished and admired by the WeRateDogs community.
